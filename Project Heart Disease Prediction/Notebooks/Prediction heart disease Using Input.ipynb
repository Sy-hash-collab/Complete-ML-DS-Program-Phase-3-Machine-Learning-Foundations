## User Input Form — Heart Disease Prediction

```python

print("Enter the information")
age	= input('age')
sex	= input('sex')           # 1 = male, 0 = female
cp = input('cp')             # Chest pain type (0–3)
trestbps = input('trestbps') # Resting blood pressure
chol = input('chol')         # Serum cholesterol
fbs	= input('fbs')           # Fasting blood sugar > 120 mg/dl (1 = true; 0 = false)
restecg	= input('restecg')   # Resting electrocardiographic results (0–2)
thalach	= input('thalach')   # Max heart rate achieved
exang = input('exang')       # Exercise-induced angina (1 = yes; 0 = no)
oldpeak = input('oldpeak')   # ST depression induced by exercise
slope = input('slope')       # Slope of the ST segment (0–2)
ca = input('ca')             # Number of major vessels (0–3)
thal = input('thal')         # Thalassemia (1 = normal; 2 = fixed defect; 3 = reversible defect)
```

```python
import pandas as pd

NewFeatures = pd.DataFrame([{
    'age': age,
    'sex': sex,
    'cp': cp,
    'trestbps': trestbps,
    'chol': chol,
    'fbs': fbs,
    'restecg': restecg,
    'thalach': thalach,
    'exang': exang,
    'oldpeak': oldpeak,
    'slope': slope,
    'ca': ca,
    'thal': thal
}])

print(NewFeatures)
```

```python
Features = NewFeatures[['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
Features
```

```python
import joblib

Model_rf = joblib.load(r'Models/random_forest_model.pickel')
Scaler = joblib.load(r'Scalers/scalers.pickel')
```
