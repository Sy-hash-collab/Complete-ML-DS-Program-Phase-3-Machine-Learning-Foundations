```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
housing
```

```python
print(housing.feature_names)
```

```python
print(housing.target)
```

```python
print(housing.data)
```

### Preparing the data

```python
dataset = pd.DataFrame(housing.data, columns=housing.feature_names)
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset['Price'] = housing.target
```

```python
dataset.info()
```

### Statistical description

```python
dataset.describe()
```

```python
## check the null values
dataset.isnull().sum()
```

```python
dataset.corr()
```

#### Detecting Outliers in dataset using Boxplot

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.boxplot(data = dataset, ax=ax)
plt.savefig("boxPlot.jpg")
```

#### Split the data into independent and dependent features

```python

X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X
```

```python
y
```

#### Split the data into train and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

#### Performing normalization on dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)
```

```python
X_train_norm
```

```python
fig, ax = plt.subplots(figsize=(15,15))
sns.boxplot(data = X_train_norm, ax=ax)
plt.savefig("boxPlotTrainData.jpg")
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
regression = LinearRegression()
regression.fit(X_train_norm, y_train)
```

### Model Prediction

```python
reg_pred = regression.predict(X_test_norm)
reg_pred
```

### Calculating Error 

```python
residuals = y_test - reg_pred
```

```python
residuals
```

### Evaluating Model performance

```python
## lower error value - MSE and MAE
## higher value - r2 score and adjusted r2 score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print("Mean Squared Error (MSE):", mean_squared_error(y_test, reg_pred))
print("Mean Absolute Error (MAE):", mean_absolute_error(y_test, reg_pred))
print("R² Score:", r2_score(y_test, reg_pred))
print("Root Mean Squared Error (RMSE):", np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
score = r2_score(y_test, reg_pred)
```
